Importing the dependences

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer # converts text into feature vectors or numerical values
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and Pre-Processing

```python
#loading the data from the csv file to a pandas dataframe
movies_data = pd.read_csv('/content/movies.csv')
```

```python
#printing the first five rows of the dataframe
movies_data.head()
```

```python
# number of rows and columns in the dataframe
movies_data.shape
```

```python
#selecting the relevant features for recommendation
selecting_features = ['genres','keywords','tagline','cast','director']
print(selecting_features)
```

```python
# replacint the missing values or the null values with null string
for feature in selecting_features:
  movies_data[feature] = movies_data[feature].fillna('')  # na in fillna means nothing available
```

```python
# combining all the 5 selected features

combined_features = movies_data['genres']+' '+movies_data['keywords']+movies_data['tagline']+movies_data['cast']+movies_data['director']
```

```python
print(combined_features)
```

```python
# Converting this above text data into numerical values or to feature vectors

vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

Cosine Similarity

```python
#getting the similarity scores using cosine similarity

similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
#getting the movie name by the user
movie_name = input(' Enter Your Favourite Movie Name : ')
```

```python
#creating a list with all the movie names in the given dataset
list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
#find the close match for the movie name inserted by the user

find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# finding the index of the movie with the title

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
#getting a list of similar movies
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting the movies based on their similarity score

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
#print the name of the similar movies based on the index

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<26):
    print(i, '.',title_from_index)
    i+=1
```

Movie Recommendation System

```python
movie_name = input('Enter your favourite movie : ')

list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<26):
    print(i, '.',title_from_index)
    i+=1
```
